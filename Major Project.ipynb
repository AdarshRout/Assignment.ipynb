```python
k = df.iloc[0:21,0:1].values # Input - X , Output - Y
k          # Converting df into array
```

```python
k.ndim
```

```python
l = df.iloc[:,1].values
l
```

```python
l.ndim
```

```python
# Running a suitable algorithm (Classifier/Regressor/Clusterer)
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# FITTING THE MODEL
model.fit(k,l)
```

```python
# Predicting The Output

l_pred = model.predict(k) # Using the input values, predict the output
l_pred # PREDICTED OUTPUT VALUES
```

```python
l # Actual Output Values
```

```python
model.predict([[29]])
```

```python
m = model.coef_
m
```

```python
c = model.intercept_
c
```

```python
# y = mx + c
m*29 + c
```

```python
# Visualisation for BEST FIT LINE 
plt.scatter(k,l) #Actual Values
plt.plot(k,l_pred, c = 'maroon') #Predicted Values
plt.title('BEST FIT LINE')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
```
